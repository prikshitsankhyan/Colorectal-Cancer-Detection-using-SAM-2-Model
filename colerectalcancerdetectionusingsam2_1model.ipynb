```python
import torch, sys
sys.path.insert(0, '/content/segment-anything-2')

from sam2.build_sam import build_sam2
from sam2.sam2_image_predictor import SAM2ImagePredictor

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

sam2_model = build_sam2(
    'configs/sam2.1/sam2.1_hiera_b+.yaml',
    '/content/checkpoints/sam2.1_hiera_base_plus.pt',
    device=DEVICE
)

predictor = SAM2ImagePredictor(sam2_model)
print("Model ready on:", DEVICE)
```

```python
import cv2, numpy as np, matplotlib.pyplot as plt
from google.colab import files

# Upload one image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Load image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Set image in predictor
predictor.set_image(image)

# Use full image as bounding box prompt
h, w = image.shape[:2]
box  = np.array([[0, 0, w, h]])

# Predict
masks, scores, _ = predictor.predict(
    point_coords=None,
    point_labels=None,
    box=box,
    multimask_output=True
)

# Pick best mask and convert to boolean
best_idx  = np.argmax(scores)
best_mask = masks[best_idx].astype(bool)  # fix is here

# Visualize
overlay            = image.copy()
overlay[best_mask] = [0, 230, 120]
blended            = cv2.addWeighted(image, 0.6, overlay, 0.4, 0)

plt.figure(figsize=(14, 4))
plt.subplot(1,3,1); plt.imshow(image);                   plt.title('Original');                          plt.axis('off')
plt.subplot(1,3,2); plt.imshow(best_mask, cmap='gray');  plt.title(f'Mask (score={scores[best_idx]:.2f})'); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(blended);                 plt.title('Overlay');                           plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import os, cv2, numpy as np, matplotlib.pyplot as plt

image_folder = '/content/drive/MyDrive/polyp_data/Train dataset/images'
mask_folder  = '/content/drive/MyDrive/polyp_data/Train dataset/masks'

# Get sorted lists so images and masks match
images = sorted(os.listdir(image_folder))
masks  = sorted(os.listdir(mask_folder))

# Check first pair matches
print("Image:", images[0])
print("Mask: ", masks[0])

# Load first pair
image    = cv2.imread(os.path.join(image_folder, images[0]))
image    = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gt_mask  = cv2.imread(os.path.join(mask_folder, masks[0]), cv2.IMREAD_GRAYSCALE)
gt_mask  = (gt_mask > 127).astype(np.uint8)

print("Image shape:", image.shape)
print("Mask shape: ", gt_mask.shape)
print("Mask unique values:", np.unique(gt_mask))

# Extract bounding box from GT mask
rows = np.any(gt_mask, axis=1)
cols = np.any(gt_mask, axis=0)
y1, y2 = np.where(rows)[0][[0, -1]]
x1, x2 = np.where(cols)[0][[0, -1]]
box = np.array([[x1, y1, x2, y2]])
print("Polyp box:", box)

# Predict
predictor.set_image(image)
masks_pred, scores, _ = predictor.predict(
    point_coords=None,
    point_labels=None,
    box=box,
    multimask_output=True
)

best_idx  = np.argmax(scores)
best_mask = masks_pred[best_idx].astype(bool)

# Visualize
overlay            = image.copy()
overlay[best_mask] = [0, 230, 120]
blended            = cv2.addWeighted(image, 0.6, overlay, 0.4, 0)

plt.figure(figsize=(16, 4))
plt.subplot(1,4,1); plt.imshow(image);                   plt.title('Original');                            plt.axis('off')
plt.subplot(1,4,2); plt.imshow(gt_mask, cmap='gray');    plt.title('Ground truth');                        plt.axis('off')
plt.subplot(1,4,3); plt.imshow(best_mask, cmap='gray');  plt.title(f'Predicted (score={scores[best_idx]:.2f})'); plt.axis('off')
plt.subplot(1,4,4); plt.imshow(blended);                 plt.title('Overlay');                             plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import os, cv2, numpy as np
from tqdm.notebook import tqdm

image_folder = '/content/drive/MyDrive/polyp_data/Train dataset/images'
mask_folder  = '/content/drive/MyDrive/polyp_data/Train dataset/masks'

images = sorted(os.listdir(image_folder))
masks  = sorted(os.listdir(mask_folder))

dice_scores = []
iou_scores  = []
failed      = []

for img_name, mask_name in tqdm(zip(images, masks), total=len(images)):

    try:
        # Load image and mask
        image   = cv2.imread(os.path.join(image_folder, img_name))
        image   = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        gt_mask = cv2.imread(os.path.join(mask_folder, mask_name), cv2.IMREAD_GRAYSCALE)
        gt_mask = (gt_mask > 127).astype(np.uint8)

        # Skip if mask is empty
        if gt_mask.sum() == 0:
            continue

        # Get bounding box from GT mask
        rows = np.any(gt_mask, axis=1)
        cols = np.any(gt_mask, axis=0)
        y1, y2 = np.where(rows)[0][[0, -1]]
        x1, x2 = np.where(cols)[0][[0, -1]]
        box = np.array([[x1, y1, x2, y2]])

        # Predict
        predictor.set_image(image)
        masks_pred, scores, _ = predictor.predict(
            point_coords=None,
            point_labels=None,
            box=box,
            multimask_output=True
        )

        best_mask = masks_pred[np.argmax(scores)].astype(np.uint8)

        # Resize predicted mask to match GT if needed
        if best_mask.shape != gt_mask.shape:
            best_mask = cv2.resize(best_mask,
                                   (gt_mask.shape[1], gt_mask.shape[0]),
                                   interpolation=cv2.INTER_NEAREST)

        # Compute Dice
        inter = (best_mask * gt_mask).sum()
        dice  = (2 * inter) / (best_mask.sum() + gt_mask.sum() + 1e-6)

        # Compute IoU
        union = (best_mask + gt_mask).clip(0, 1).sum()
        iou   = inter / (union + 1e-6)

        dice_scores.append(dice)
        iou_scores.append(iou)

    except Exception as e:
        failed.append(img_name)

# Final results
print("=" * 40)
print(f"Total images evaluated : {len(dice_scores)}")
print(f"Failed                 : {len(failed)}")
print(f"mDice                  : {np.mean(dice_scores):.4f}")
print(f"mIoU                   : {np.mean(iou_scores):.4f}")
print(f"Best  Dice             : {np.max(dice_scores):.4f}")
print(f"Worst Dice             : {np.min(dice_scores):.4f}")
print("=" * 40)
```
